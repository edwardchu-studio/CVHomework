```python
def displayImage(curImage):
    plt.figure(figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

    plt.subplot(111)
    plt.imshow(curImage,cmap='gray')
    plt.tight_layout()
    
def rgb2gray(rgb):
    if len(rgb.shape)<=2:
        return np.array(rgb)
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

### Problem 1 Affine Image Transformation
##### Problem1.1 Resampling

```python
def getTranslationMetrix(shift=(0,0)):
    m=np.array([
        [1,0,shift[0]],
        [0,1,shift[1]],
        [0,0,1]
    ])
    return m
def getRotateMetrix(theta=0):
    m=np.array([
        [np.cos(theta),-np.sin(theta),0],
        [np.sin(theta),np.cos(theta),0],
        [0,0,1]
    ])
    return m
def getScaleMetrix(scale=(1,1)):
    m=np.array([
        [scale[0],0,0],
        [0,scale[1],0],
        [0,0,1]
    ])
    return m
def getShearMetrix(shear=(0,0)):
    m=np.array([
        [1,shear[1],0],
        [shear[0],1,0],
        [0,0,1]
    ])
    return m
def transformByNN(raw,m):
    h,w=raw.shape
    new=raw.copy()
    for i in range(h-1,-1,-1):
        for j in range(w-1,-1,-1):
            x_,y_,_=m.dot([i,j,1]).astype(int)
            try:
                new[x_,y_]=img1[i,j]
            except:
#                 print(i,j,x_,y_)
                x_,y_=x_%h,y_%w
#                 new[x_,y_]=raw[i,j]
                new[x_,y_]=255
    return new

def compTrans(raw,m,t, sample='nn'):
    plt.figure(figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

    plt.subplot(121)
    plt.imshow(raw,cmap='gray')
    plt.title(s='Raw Image')

    new=transformByNN(raw,m) if sample=='nn' else transformByNN(raw,m)
    plt.subplot(122)
    plt.imshow(new,cmap='gray')
    plt.title(s=t+' Image')
    plt.tight_layout()
    
def transformByBilinear(raw,m):
    pass
```

```python
compTrans(img1,getTranslationMetrix((20,15)),t='Translated')
compTrans(img1,getRotateMetrix(theta=np.pi/16),t='Rotated')
compTrans(img1,getScaleMetrix((1.2,1.2)),t='Scaled')
compTrans(img1,getShearMetrix((-0.2,0)),t='Sheared')
```

### Problem 2 Hough Trnasform for Straight Lines
##### Problem2.1  Without Edge Orient

```python
def generateSketch(theta,rous):
    w=len(theta)
    h=rous.max()-rous.min()
    return np.zeros((int(h),w))

def detectLine(img):
    ratio=0.8
    edge= feature.canny(img)
    theta=np.linspace(start=-np.pi/2,stop=np.pi/2,num=600)
    imgH,imgW=edge.shape
    dots=[(x,imgH-y) for y in range(edge.shape[0]) for x  in range(edge.shape[1]) if edge[y,x]>0]
    rous=[]
    for x,y in dots:
        rous.append(x*np.cos(theta)+y*np.sin(theta))
    rous=np.array(rous)
    sketch=generateSketch(theta,rous)
    bottom=rous.min()
    h,w=sketch.shape
    for rou in rous:
        for t in range(len(theta)):
            sketch[h-(int(rou[t]-bottom)+1)][t]+=1
    displayImage(sketch)
    num_lines=int(sketch.max()*(1-ratio))
    lines=[]
    while(num_lines):
        ind = np.unravel_index(np.argmax(sketch, axis=None), sketch.shape)
        num_lines-=1
        sketch[ind]=0
        t=-np.pi/2+np.pi*(ind[1])/600
        r=rous[ind]
        lines.append((t,r))
    return lines
```

```python
def fun():
    lines=detectLine(img2)
    print("{} lines detected, with following parameters:".format(len(lines)))
    ab=[]
    for t,r in lines:
        ab.append((-np.tan(t),r/np.sin(t)))
        print("theta:{},rou:{}, calculated line is:[y={}x+{}]".format(t,r,-np.tan(t),r/np.sin(t)))
    x=np.linspace(num=100,start=0,stop=img1.shape[1])
    plt.figure()
    plt.subplot(121)
    plt.imshow(img2,cmap='gray')
    plt.subplot(122)
    h,w=img2.shape
    plt.axis([0,w,0,h])
#     ax = plt.gca()
#     ax.set_autoscale_on(False)
    for a,b in ab:
        plt.plot(a*x+b)
fun()
```
